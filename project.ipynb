#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
south_belt_df['precio_normalizado'] = south_belt_df.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())
plt.figure(figsize=(10, 6))
bins = 20
for población in south_belt:
    subset = south_belt_df[south_belt_df['level5'] == población]
    plt.hist(subset['precio_normalizado'], bins=bins, alpha=0.7, label=población)
plt.title('Histograma de precios normalizados por población', fontsize=16)
plt.xlabel('Precios Normalizados', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Población')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
